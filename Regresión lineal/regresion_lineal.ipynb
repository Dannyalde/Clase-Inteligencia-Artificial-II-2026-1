```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np


#Datos 
x = [173, 171, 189, 181] 
y = [81, 72, 96, 94]


# Convertir a arrays de numpy y reshape para sklearn
X = np.array(x).reshape(-1,1) # fila a columna
Y = np.array(y)


# crea y se entrena el modelo 
modelo = LinearRegression() 
modelo.fit(X,Y)


# obtener parametros del modelo
b = modelo.intercept_
m = modelo.coef_


# obtener prediccion
y_pred = modelo.predict(X)
print(y_pred)
print(Y)

# Calcular m√©tricas
r2 = r2_score(Y, y_pred)
mse = mean_squared_error(Y, y_pred)
print(r2, mse)

print(f" Punto de corte: {b}")
print(f"Pendiente: {m}")


```

```python
# Crear visualizaci√≥n
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', s=100, label='Datos reales')
plt.plot(x, y_pred, color='red', linewidth=2, label=f'Regresi√≥n lineal (R¬≤={r2:.4f})')

plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Peso (kg)', fontsize=12)
plt.title('Regresi√≥n Lineal: Predicci√≥n de Peso basado en Altura', fontsize=14, fontweight='bold')
plt.legend()
plt.grid()
#plt.tight_layout()
plt.legend()
```

```python

#Datos 
x = np.array(data["GrLivArea"]).reshape(-1,1)
y = np.array(data["SalePrice"])


# crea y se entrena el modelo 
modelo = LinearRegression() 
modelo.fit(x,y)


# obtener parametros del modelo
b = modelo.intercept_
m = modelo.coef_


# obtener prediccion
y_pred = modelo.predict(x)
print(y_pred)
print(y)

# Calcular m√©tricas
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
print(r2, (mse)**(1/2))

print(f" Punto de corte: {b}")
print(f"Pendiente: {m}")







```
